### 1. Membuat gelombang sinus dan Cosinus

```python
import numpy as np
import matplotlib.pyplot as plt

# Buat array t yang berkisar dari 0 hingga 2 dengan langkah sebesar 0.0001
t = np.arange(0, 2, 0.0001)

# Buat sinyal-sinyal menggunakan array waktu
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)

```

### 2. Perbandingan subplot

```python
# Membuat plot dengan 4 subplot dalam grid 2x2
fig, axs = plt.subplots(2, 2, figsize=(6, 5))
fig.suptitle('Perbandingan Sinyal Sinus dan Cosinus', fontsize=16)

# Plot untuk y1
axs[0, 0].plot(t, y1)
axs[0, 0].set_title('y1 = 2 * sin(2 * 3t + 0)')
axs[0, 0].set_ylabel('Amplitudo')

# Plot untuk y2
axs[0, 1].plot(t, y2)
axs[0, 1].set_title('y2 = 1 * cos(2 * 4t + /4)')
axs[0, 1].set_ylabel('Amplitudo')

# Plot untuk y3
axs[1, 0].plot(t, y3)
axs[1, 0].set_title('y3 = -1 * sin(2 * 5t + /2)')
axs[1, 0].set_xlabel('Waktu')
axs[1, 0].set_ylabel('Amplitudo')

# Plot untuk y4
axs[1, 1].plot(t, y4)
axs[1, 1].set_title('y4 = 0.5 * cos(2 * 6t + )')
axs[1, 1].set_xlabel('Waktu')
axs[1, 1].set_ylabel('Amplitudo')

# Mengatur layout agar tidak tumpang tindih
plt.tight_layout()
plt.show()
```

#### - bagaimana kombinasi dua gelombang sinus/cosinus mempengaruhi bentuk gelombang yang dihasilkan.

Hasil kombinasi ini akan menghasilkan gelombang yang memiliki ;

- Bentuk yang tidak beraturan: Karena frekuensi yang berbeda, yaitu y1 (3 Hz) memiliki periode yang lebih panjang dibandingkan y2 (4 Hz). gelombang akan memiliki pola yang berulang setelah periode tertentu (ketika kedua komponen kembali sefase). y5 akan memiliki bentuk yang lebih kompleks dibandingkan gelombang sinus atau cosinus sederhana.

- Amplitudo yang bervariasi: Pada beberapa titik, kedua gelombang akan saling menguatkan (interferensi konstruktif), sementara pada titik lain akan saling melemahkan (interferensi destruktif). sinyal y5 akan menunjukkan variasi amplitudo yang lebih besar dari komponen-komponennya.

- Karena perbedaan frekuensi yang kecil (4 Hz - 3 Hz = 1 Hz), akan terlihat pola "beating" atau modulasi amplitudo dengan frekuensi yang lebih rendah yang merupakan hasil dari perbedaan frekuensi antara y1 dan y2..

- Asimetri: Kombinasi sinus dan cosinus akan menghasilkan gelombang yang tidak simetris terhadap sumbu waktu.

### 5. downsampling dari 150Hz ke 100Hz, 50Hz, 25Hz, 10Hz, hingga 5Hz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import neurokit2 as nk

# Membuat sinyal ECG sintetis
ecg = nk.ecg_simulate(duration=92, sampling_rate=150, noise=0.92, heart_rate=80, random_state=30819)

# Fungsi untuk melakukan downsampling
def downsample(signal, original_fs, target_fs):
    factor = original_fs // target_fs
    return signal[::factor]

# Fungsi untuk plot sinyal
def plot_signal(signal, fs, title):
    time = np.arange(len(signal)) / fs
    plt.figure(figsize=(12, 4))
    plt.plot(time, signal)
    plt.title(title)
    plt.xlabel('Waktu (detik)')
    plt.ylabel('Amplitudo')
    plt.grid(True)
    plt.show()

# Plot sinyal asli
plot_signal(ecg, sampling_rate, 'Sinyal ECG Asli (150 Hz)')

# Lakukan downsampling dan plot hasilnya
target_rates = [100, 50, 25, 10, 5]

for rate in target_rates:
    downsampled = downsample(ecg, sampling_rate, rate)
    plot_signal(downsampled, rate, f'Sinyal ECG Downsampled ({rate} Hz)')

# Demonstrasi aliasing
def demonstrate_aliasing():
    t = np.linspace(0, 1, 1000)
    original_signal = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 20 * t)
    
    plt.figure(figsize=(12, 8))
    
    plt.subplot(3, 1, 1)
    plt.plot(t, original_signal)
    plt.title('Sinyal Asli')
    plt.xlabel('Waktu')
    plt.ylabel('Amplitudo')
    
    # Downsampling yang menyebabkan aliasing
    aliased_signal = original_signal[::20]
    t_aliased = t[::20]
    
    plt.subplot(3, 1, 2)
    plt.plot(t_aliased, aliased_signal)
    plt.title('Sinyal Ter-aliasing')
    plt.xlabel('Waktu')
    plt.ylabel('Amplitudo')
    
    # Rekonstruksi sinyal (menunjukkan efek aliasing)
    reconstructed = np.interp(t, t_aliased, aliased_signal)
    
    plt.subplot(3, 1, 3)
    plt.plot(t, reconstructed)
    plt.title('Sinyal Hasil Rekonstruksi (Menunjukkan Efek Aliasing)')
    plt.xlabel('Waktu')
    plt.ylabel('Amplitudo')
    
    plt.tight_layout()
    plt.show()

demonstrate_aliasing()
```

Aliasing adalah suatu efek yang terjadi jika tidak memenuhi kriteria maka akan timbul frekuensi-frekuensi yang tidak diinginkan yang tidak terdapat pada sinyal asli sehingga sinyal asli tidak akan didapatkan kembali tanpa cacat. (Prasetiyo, Sofyan Mufti, et al. "Teknologi Komunikasi Digital Dan Analog: Konversi, Transmisi." Buletin Ilmiah Ilmu Komputer dan Multimedia (BIIKMA) 1.6 (2024): 750-762.)

Efek Penurunan Sampling Frequency:
1. 150 Hz (Sinyal Asli):
- sinyal ECG dengan kualitas terbaik
- Semua komponen frekuensi sampai 75 Hz (setengah dari sampling rate) dapat direpresentasikan dengan akurat.

2. 100 Hz:
- Masih cukup akurat untuk sebagian besar aplikasi ECG.
- Komponen frekuensi di atas 50 Hz mulai kehilangan representasi yang akurat.

3. 50 Hz:
- terlihat sedikit distorsi pada bagian dengan perubahan cepat seperti kompleks QRS.
- Komponen frekuensi di atas 25 Hz tidak lagi direpresentasikan dengan akurat.

4. 25 Hz:
- Distorsi menjadi lebih jelas.
- Detail halus pada gelombang P dan T mulai hilang.
- Bentuk kompleks QRS menjadi kurang tajam.

5. 10 Hz:
- Distorsi sangat signifikan.
- Banyak detail penting hilang.
- Bentuk gelombang menjadi sangat kasar dan tidak akurat.

- Sinyal hampir tidak dapat dikenali sebagai ECG.

Semakin rendah sampling frequency (fs) maka sinyal akan semakin terdistorsi dan terdapat Aliasing pada sinyal hasil downsampling.
hal ini dibuktikan dengan adanya 
- distorsi bentuk gelombang: Semakin rendah fs, semakin tidak akurat bentuk gelombang ECG yang dihasilkan, terutama pada kompleks QRS.
- Hilangnya Detail: Pada fs yang sangat rendah (10 Hz, 5 Hz), banyak detail penting dari sinyal ECG yang hilang.
- Kemunculan Aliasing: Pada demonstrasi aliasing, terlihat jelas bagaimana sinyal frekuensi tinggi "menyamar" menjadi sinyal frekuensi rendah ketika disampling terlalu rendah.


### 6. Parameter order saat filtering

Parameter order pada pemrosesan sinyal yaitu tingkat kompleksitas atau tingkat kebebasan sebuah filter atau model. Semakin tinggi order suatu filter, semakin kompleks bentuk respon frekuensinya, yang artinya filter tersebut mampu melakukan operasi yang lebih rumit pada sinyal.

Dampak jika mengubah nilai order yaitu:

- Kompleksitas Perhitungan: Semakin tinggi order, semakin kompleks perhitungan yang diperlukan untuk mendesain dan mengimplementasikan filter atau model.
- Akurasi Model: Order yang terlalu rendah dapat menyebabkan underfitting, di mana model tidak mampu menangkap kompleksitas data. Sebaliknya, order yang terlalu tinggi dapat menyebabkan overfitting, di mana model terlalu menyesuaikan diri dengan noise dalam data.
- Biaya Komputasi: Filter dengan order tinggi umumnya membutuhkan lebih banyak sumber daya komputasi untuk diimplementasikan secara real-time.
- Respon Frekuensi: Order filter mempengaruhi bentuk respon frekuensi, seperti tingkat riak, ketajaman transisi, dan selektivitas frekuensi.

### 7. eksperimen dengan merancang filter band-pass menggunakan signal.butter

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

fs = 1000  # Frekuensi sampling dalam Hz
lowcut = 80  # Frekuensi cut-off bawah dalam Hz
highcut = 200  # Frekuensi cut-off atas dalam Hz
order = 7 

# Fungsi untuk membuat filter band-pass
def butter_bandpass(lowcut, highcut, fs, order):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='band')
    return b, a

# Fungsi untuk menerapkan filter
def butter_bandpass_filter(data, lowcut, highcut, fs, order):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    y = signal.lfilter(b, a, data)
    return y

# Buat sinyal contoh
t = np.linspace(0, 1, fs, False)  # 1 detik
signal_clean = np.sin(2*np.pi*100*t) + 0.5*np.sin(2*np.pi*300*t)
signal_noisy = signal_clean + 2*np.random.randn(len(t))

# Terapkan filter
filtered_signal = butter_bandpass_filter(signal_noisy, lowcut, highcut, fs, order)

# Plot hasil
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, signal_clean)
plt.title('Sinyal Asli')
plt.xlabel('Waktu [detik]')
plt.ylabel('Amplitudo')

plt.subplot(3, 1, 2)
plt.plot(t, signal_noisy)
plt.title('Sinyal dengan Noise')
plt.xlabel('Waktu [detik]')
plt.ylabel('Amplitudo')

plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal)
plt.title('Sinyal Setelah Filter Band-Pass')
plt.xlabel('Waktu [detik]')
plt.ylabel('Amplitudo')

plt.tight_layout()
plt.show()

# Plot respons frekuensi filter
w, h = signal.freqs(*butter_bandpass(lowcut, highcut, fs, order))
plt.figure(figsize=(5, 3))
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Respons Frekuensi Filter Band-Pass Butterworth')
plt.xlabel('Frekuensi [Hz]')
plt.ylabel('Amplitudo [dB]')
plt.grid(True)
plt.axvline(lowcut, color='green')  # Frekuensi cut-off bawah
plt.axvline(highcut, color='green')  # Frekuensi cut-off atas
plt.show()
```
