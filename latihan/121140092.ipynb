```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time axis
time_axis = np.linspace(0, 2 * np.pi, 1000)

# Define the input signals
input_signal1 = np.sin(time_axis)
input_signal2 = np.cos(time_axis)

# Plot both signals
plt.plot(time_axis, input_signal1, label='Sine Wave')
plt.plot(time_axis, input_signal2, label='Cosine Wave')
plt.title('Sine and Cosine Waves')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistem yang hanya mengembalikan sinyal yang diterimanya
def sistem(signal):
    return 2 * signal  # Misalkan sistem kita mengalikan sinyal dengan 2

# Define the time axis
time_axis = np.linspace(0, 2 * np.pi, 1000)

# Define the input signals
input_signal1 = np.sin(time_axis)
input_signal2 = np.cos(time_axis)

# Perbesar sinyal dengan faktor 3
input_signal1_scaled = 3 * input_signal1
input_signal2_scaled = 3 * input_signal2

# Kirim sinyal yang sudah diperbesar ke sistem
output_signal1 = sistem(input_signal1_scaled)
output_signal2 = sistem(input_signal2_scaled)

# Plot hasil perbesaran dan output sistem
plt.figure(figsize=(10, 6))

# Plot sinyal input setelah perbesaran
plt.subplot(2, 1, 1)
plt.plot(time_axis, input_signal1_scaled, label='Scaled Sine Wave (3x)', color='blue')
plt.plot(time_axis, input_signal2_scaled, label='Scaled Cosine Wave (3x)', color='red')
plt.title('Scaled Input Signals (3x Magnification)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Plot output dari sistem
plt.subplot(2, 1, 2)
plt.plot(time_axis, output_signal1, label='System Output (Sine)', color='blue')
plt.plot(time_axis, output_signal2, label='System Output (Cosine)', color='red')
plt.title('System Output after Scaling')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistem yang mengalikan sinyal dengan 2
def sistem(signal):
    return 2 * signal

# Define the time axis
time_axis = np.linspace(0, 2 * np.pi, 1000)

# Define the input signals
input_signal1 = np.sin(time_axis)
input_signal2 = np.cos(time_axis)

# Penjumlahan sebelum dikirim ke sistem
input_sum = input_signal1 + input_signal2
output_sum_before = sistem(input_sum)

# Kirimkan sinyal 1 dan 2 ke sistem secara terpisah, lalu jumlahkan hasilnya
output_signal1 = sistem(input_signal1)
output_signal2 = sistem(input_signal2)
output_sum_after = output_signal1 + output_signal2

# Plot hasilnya untuk visualisasi
plt.figure(figsize=(10, 6))

# Plot penjumlahan sebelum sistem
plt.subplot(2, 1, 1)
plt.plot(time_axis, output_sum_before, label='Sum Before System', color='blue')
plt.title('Penjumlahan Sebelum Sistem')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

# Plot penjumlahan setelah sistem
plt.subplot(2, 1, 2)
plt.plot(time_axis, output_sum_after, label='Sum After System', color='red')
plt.title('Penjumlahan Setelah Sistem')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

# Cek apakah hasilnya sama
if np.allclose(output_sum_before, output_sum_after):
    print("Penjumlahan sebelum dan sesudah sistem menghasilkan hasil yang sama.")
else:
    print("Penjumlahan sebelum dan sesudah sistem menghasilkan hasil yang berbeda.")
```
